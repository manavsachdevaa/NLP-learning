# NLTK -> NATURAL LANGUAGE TOOLKIT

```python
import nltk
#4 things to do with NLTK
#a. Tokenization
#b. Stopword Removal
#c. Stemming
#d. Lemmetization
```

```python
nltk.download("punkt")     #punkt-> punctuation
nltk.download('punkt_tab')
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
#TOkennization
from nltk.tokenize import word_tokenize
#Stopwords List
from nltk.corpus import stopwords
#Stemming
from nltk.stem import PorterStemmer
#Lemmatization
from nltk.stem import WordNetLemmatizer
```

```python
raw_text="I am learning Python Programming, Python is a great language...!!"
```

```python
#Step1-> : Lowercase
raw_text=raw_text.lower()
print("After lowercase: ",raw_text)
```

```python
#Step2-> Tokenization
print(raw_text.split())
```

```python
tokens=word_tokenize(raw_text)
print("After Tokenization: ",tokens)
```

```python
#Remove stopwords
english_stopwords=stopwords.words("english")
print("English Stopwords: ",english_stopwords)
```

```python
filtered_tokens=[]
for token in tokens:
    if token not in english_stopwords:
        filtered_tokens.append(token)
print("Filtered Tokens :",filtered_tokens)
```

```python
#Step-4 Remove punctuations
import string
punctuations=string.punctuation
print("Punctuations Available: ",punctuations)
```

```python
clean_tokens=[]
for token in filtered_tokens:
    if token not in punctuations:
        clean_tokens.append(token)
print("After removing punctuations: ",clean_tokens)
```

```python
stem=PorterStemmer()
# stem.stem("learning")
# stem.stem("playing")
# stem.stem("flying")
stem.stem("bought")
```

```python
wnet=WordNetLemmatizer()
# wnet.lemmatize("playing","v")
# wnet.lemmatize("flying","v")
# wnet.lemmatize("bought","v")
wnet.lemmatize("went","v")
```

```python
lemmtized_words=[]
for word in clean_tokens:
    lemmtized_words.append(wnet.lemmatize(word,"v"))
print("After Lemmatization: ",lemmtized_words)
```

```python
final_tokens=[]
for word in lemmtized_words:
  if word.isalpha():
    final_tokens.append(word)
print("Final Tokens: ",final_tokens)
```

```python
cleaned_text=" ".join(final_tokens)
print("Cleaned text: ",cleaned_text)
```

```python
#Function

sentences=[
    "i am Learning Python and python" ,
    "we are happy today",
    "i am sad today"
]


def preprocess_text(documents):
  final_list=[]
  for sent in documents:
    sent=sent.lower()
    tokens=word_tokenize(sent)
    english_stopwords=stopwords.words("english")
    filtered_tokens=[]

    for token in tokens:
      if token not in english_stopwords:
        filtered_tokens.append(token)
    punctuations=string.punctuation
    clean_tokens=[]
    for token in filtered_tokens:
      if token not in punctuations:
        clean_tokens.append(token)
    wnet=WordNetLemmatizer()
    lemmtized_words=[]
    for word in clean_tokens:
      lemmtized_words.append(wnet.lemmatize(word,"v"))
    alpha_list=[]
    for word in lemmtized_words:
      if word.isalpha():
        alpha_list.append(word)
    final_list.append(" ".join(alpha_list))
  return final_list









cleaned_sentences=preprocess_text(sentences)
print("Cleaned Sentences: ",cleaned_sentences)
```
