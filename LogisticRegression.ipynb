# NPL-BASED CHATBOT

```python
#TOkennization
from nltk.tokenize import word_tokenize
#Stopwords List
from nltk.corpus import stopwords
#Stemming
from nltk.stem import PorterStemmer
#Lemmatization
from nltk.stem import WordNetLemmatizer
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
import nltk
nltk.download("punkt")     #punkt-> punctuation
nltk.download('punkt_tab')
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
training_data = [

    # ---------- GREET (9) ----------
    ("hello", "greet"),
    ("hi", "greet"),
    ("hey", "greet"),
    ("hello there", "greet"),
    ("hi there", "greet"),
    ("good morning", "greet"),
    ("good evening", "greet"),
    ("hey buddy", "greet"),
    ("hii", "greet"),

    # ---------- WEATHER (9) ----------
    ("what is the weather today", "weather"),
    ("how is the weather today", "weather"),
    ("tell me the weather", "weather"),
    ("what's the temperature today", "weather"),
    ("how is the temperature", "weather"),
    ("is it raining", "weather"),
    ("will it rain today", "weather"),
    ("tell me temperature", "weather"),
    ("weather update", "weather"),

    # ---------- OPEN_WEB (9) ----------
    ("open google", "open_web"),
    ("open facebook", "open_web"),
    ("open youtube", "open_web"),
    ("open instagram", "open_web"),
    ("open twitter", "open_web"),
    ("open linkedin", "open_web"),
    ("open gmail", "open_web"),
    ("go to google", "open_web"),
    ("launch youtube", "open_web"),

    # ---------- EXIT (8) ----------
    ("bye", "exit"),
    ("goodbye", "exit"),
    ("bye bye", "exit"),
    ("see you later", "exit"),
    ("exit", "exit"),
    ("quit", "exit"),
    ("close the app", "exit"),
    ("stop", "exit"),
]

```

```python
sentences=[]   #features / x
labels=[]      #target / y

for text,intent in training_data:
    sentences.append(text)
    labels.append(intent)
```

```python
sentences[:5]
```

```python
def preprocess_text(documents):
  final_list=[]
  for sent in documents:
    sent=sent.lower()
    tokens=word_tokenize(sent)
    english_stopwords=stopwords.words("english")
    filtered_tokens=[]

    for token in tokens:
      if token not in english_stopwords:
        filtered_tokens.append(token)
    punctuations=string.punctuation
    clean_tokens=[]
    for token in filtered_tokens:
      if token not in punctuations:
        clean_tokens.append(token)
    wnet=WordNetLemmatizer()
    lemmtized_words=[]
    for word in clean_tokens:
      lemmtized_words.append(wnet.lemmatize(word,"v"))
    alpha_list=[]
    for word in lemmtized_words:
      if word.isalpha():
        alpha_list.append(word)
    final_list.append(" ".join(alpha_list))
  return final_list
```

```python
cleaned_data=preprocess_text(sentences)
cleaned_data
```

```python
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(cleaned_data)    #raining   (Words -> Vector)

```

```python
model=LogisticRegression()
model.fit(X,labels)      #X and Y
```

```python
user_msg=""  #Testing data
processed=preprocess_text([user_msg])
user_vector=vectorizer.transform(processed)    #For testing only transform for training fit_transform
prediction=model.predict(user_vector)
print("Prediction is: ",prediction)
```

```python
print(user_vector.toarray())
print(model.intercept_)
print(model.classes_)
```
